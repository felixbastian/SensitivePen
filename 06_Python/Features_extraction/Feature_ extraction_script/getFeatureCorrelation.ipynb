```python
import numpy as np
import pandas as pd
```

```python
# Set data path and import data
data_path = "Data_summary.xlsx"
df_data = pd.read_excel(data_path).drop(columns=["Unnamed: 0"])
```

```python
df_data.columns
```

```python
# List features
features = ["AZ_mean","AZ_max","AZ_min","AZ_std","AZ_var","AL_mean","AL_max","AL_min","AL_std","AL_var",
           "AZ_diff_mean","AZ_diff_max","AZ_diff_min","AZ_diff_std","AZ_diff_var","AL_diff_mean","AL_diff_max",
           "AL_diff_min","AL_diff_std","AL_diff_var","AL_percentile_60","acceleration_mean","acceleration_max",
           "acceleration_min","acceleration_D","jerk_mean","jerk_max","jerk_min","jerk_std","jerk_mad","jerk_skew",
           "jerk_entropy","nb_max_acc","nb_min_acc"]

labels = ["BHK_speed","BHK_quality"]
```

### Feature to Feature Correlation

```python
# Analyze Feature Pearson Correlation
df_data_feats = df_data[features]
feature_corr = df_data_feats.corr(method="pearson").abs()
feature_corr.style.background_gradient(cmap="coolwarm")
```

```python
# Remove pairs of same column and sort with regards to correlation
pairs_to_drop = set((column,column) for column in features)
sorted_corr = feature_corr.unstack().drop(labels=pairs_to_drop).sort_values(ascending=False)
```

```python
# Plot n most and least correlated features
n=10

print(n,"most correlated features")
print(sorted_corr[:n])
print("")
print(n,"least correlated features")
print(sorted_corr[-n:])
```

```python
corr_sum = (feature_corr.sum(axis=1) - 1).sort_values(ascending=False)
print("Features with highest correlation to others")
print()
print(corr_sum)
```

```python
corr_std = feature_corr.std(axis=1).sort_values(ascending=False)
print("Variance in correlation per feature")
print()
print(corr_std)
```

### Feature to Label Correlation

```python
# Create feature to label correlation matrix
feat_lbl_corr = df_data.corr(method='pearson').drop(index=labels,columns=features).abs()
feat_lbl_corr.style.background_gradient(cmap="coolwarm")
```

```python
# Rank features with reagards to their speed correlation
print("Ranked Features - Correlation to BHK Speed")
feat_lbl_corr["BHK_speed"].sort_values(ascending=False)
```

```python
# Rank features with reagards to their speed correlation
print("Ranked Features - Correlation to BHK Quality")
feat_lbl_corr["BHK_quality"].sort_values(ascending=False)
```

```python
# Rank features with regards to the sum of their correlations
feat_lbl_corr["BHK_total"] = feat_lbl_corr["BHK_speed"] + feat_lbl_corr["BHK_quality"]
feat_lbl_corr.sort_values("BHK_total", ascending=False).style.background_gradient(cmap="coolwarm")
```
